# Clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
model1 = KMeans(n_clusters=4, random_state=101)
```

```python
model1.fit(X)
```

```python
pred = model1.predict(X)
```

```python
df["Cluster"] = pred
```

```python
df.sample(5)
```

```python
silhouette_score(X,pred)
```

```python
pred
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(df['Military Strength Power Index'],df['Aircraft Strength value'],c=pred)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Navy Fleet Strengths value'],df['Aircraft Strength value'],c=pred);
```

```python
index = 0
plt.figure(figsize=(25, 60))

for feature in df.drop("Country", axis=1):
    if feature != "Cluster":
        index += 1
        plt.subplot(9, 3, index)
        sns.boxplot(x="Cluster", y=feature, data=df)
```

```python
allscore=[]
allcluster=[]
distances=[]

def clust(k):
    for i in range(1,k):
        i=i+1
        model = KMeans(i)
        model = model.fit(X)
        pred = model.predict(X)
        score=silhouette_score(X,pred)
        distance = model.inertia_
        print("Cluster", i, "score", score)
        allscore.append(score)
        allcluster.append(i)
        distances.append(distance)
```

```python
df[df["Cluster"] == 1][["Country", "Military Strength Power Index"]].sort_values(by="Military Strength Power Index")
```

```python
df[df["Cluster"] == 2][["Country", "Military Strength Power Index"]].sort_values(by="Military Strength Power Index")
```

```python
df[df["Cluster"] == 3][["Country", "Military Strength Power Index"]].sort_values(by="Military Strength Power Index")
```

```python
clust(10)
```

```python
plt.plot(allcluster,distances)
```
